```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_data = pd.read_csv('car data.csv')
car_data.head()
```

# Exploratory Data Analysis

```python
ax = sns.countplot(x="Seller_Type", data=car_data)
```

```python
ax = sns.countplot(x="Fuel_Type", data=car_data)
```

```python
ax = sns.countplot(x="Owner", data=car_data)
```

```python
car_data.Car_Name = car_data.Car_Name.replace('Activa 4g', 'Honda Activa 4G')
car_data.Car_Name = car_data.Car_Name.replace('Activa 3g', 'Honda Activa 3G')
car_data.Car_Name = car_data.Car_Name.replace('KTM 390 Duke ', 'KTM RC390')
car_data.Car_Name = car_data.Car_Name.replace('Bajaj Pulsar  NS 200', 'Bajaj Pulsar NS 200')
sorted(car_data.Car_Name.unique())
```

```python
plt.figure(figsize=(10,80))
sns.countplot(y='Car_Name', data = car_data)
```
